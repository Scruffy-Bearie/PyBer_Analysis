# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = total_fares/total_rides
average_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = total_fares/total_drivers
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides":total_rides,"Total Drivers":total_drivers,"Total Fares":total_fares,"Average Fare per Ride":average_fare_ride,"Average Fare per Driver":average_fare_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].astype(int).apply(lambda x: f'{x:,}') 
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].astype(int).apply(lambda x: f'{x:,}')
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].astype(int).apply(lambda x: f'{x:,}')

pyber_summary_df.loc[:,"Total Fares"] = pyber_summary_df.loc[:,"Total Fares"].astype('str')
pyber_summary_df.loc[:,"Average Fare per Ride"] = pyber_summary_df.loc[:,"Average Fare per Ride"].round(2)
pyber_summary_df.loc[:,"Average Fare per Ride"] = pyber_summary_df.loc[:,"Average Fare per Ride"].astype(str)
pyber_summary_df.loc[:,"Average Fare per Driver"] = pyber_summary_df.loc[:,"Average Fare per Driver"].round(2)
pyber_summary_df.loc[:,"Average Fare per Driver"] = pyber_summary_df.loc[:,"Average Fare per Driver"].astype(str)

pyber_summary_df["Total Fares"] = ('$' + pyber_summary_df["Total Fares"])
pyber_summary_df["Average Fare per Ride"] = ('$' + pyber_summary_df["Average Fare per Ride"])
pyber_summary_df["Average Fare per Driver"] = ('$' + pyber_summary_df["Average Fare per Driver"])

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
npyber_data_df = pyber_data_df
npyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
npyber_fares_df = npyber_data_df.groupby(["type","date"]).sum()
nnpyber_fares_df = npyber_fares_df.loc[:,["fare"]]
nnpyber_fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
nnpyber_fares_df = nnpyber_fares_df.reset_index()
nnpyber_fares_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_pivot = nnpyber_fares_df.pivot(index = "date", columns = "type", values = "fare")
fares_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_pivot_select = fares_pivot.loc["2019-01-01":"2019-04-29"]
fares_pivot_select
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_pivot_select.index = pd.to_datetime(fares_pivot_select.index)
fares_pivot_select
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fares_pivot_select.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares = fares_pivot_select.resample("w").sum()
weekly_fares
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

import matplotlib.dates as mdates

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(25, 8))

ax.plot((weekly_fares), label= ('Rural','Suburban','Urban'))
ax.legend(loc = 'center')


ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(2,3,4)))

ax.set_ylabel('Fare ($USD)')
ax.set_title("Total Fare by City Type")

ax.annotate("Jan", (63,18), xytext=(10,10), xycoords="figure pixels", textcoords="offset pixels", arrowprops=None, annotation_clip=None)
ax.annotate("2019", (58,1), xytext=(10,10), xycoords="figure pixels", textcoords="offset pixels", arrowprops=None, annotation_clip=None)

plt.savefig("Analysis/PyBer_fare_summary.png")
```
