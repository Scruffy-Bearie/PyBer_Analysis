```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis,marker = "*", color = "green", linewidth = 2, label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.ylim(0,45)
plt.title("Fare as a Function of Month")
plt.grid()
plt.legend()
```

```python
plt.bar(x_axis,y_axis,color = "green", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Fare as a Function of Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
ax.invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, c = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.ylabel("Month")
plt.xlabel("Average Fare ($)")
plt.title("Average Fare as a Function of Month")
plt.legend()
```

```python
plt.figure(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, shadow = True, explode = explode_values, colors = colors, labels=x_axis, autopct='%.1f%%')
```
